## Q1.Total number of apps in each category

```python
import pandas as pd
df = pd.read_csv('googleplaystore.csv')
for name in df['Category'].unique():
    ct = 0
    for i in df['Category']:
        if (i == name):
            ct += 1
    print("There are total", ct, "apps in the", name, "category")
```

```python
# We can also pass the result into a dictionary format
import pandas as pd
df = pd.read_csv('googleplaystore.csv')

categories ={} #creating a dictionary

for name in df['Category'].unique():
    ct = 0
    for i in df['Category']:
        if (i == name):
            ct += 1
    categories[name] = ct
print(categories)
```

## Q2.Total number of apps in each Type

```python
types ={} #creating a dictionary

for name in df['Type'].unique():
    ct = 0
    for i in df['Type']:
        if (i == name):
            ct += 1
    types[name] = ct
    print(types)
```
